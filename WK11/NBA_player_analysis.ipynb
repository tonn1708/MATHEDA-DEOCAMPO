## Team data analysis

```python
drafted_players = df[df['draft_round'] != 'Undrafted']
```

```python
players_team_drafted= drafted_players['team_abbreviation'].value_counts()
players_team_drafted.head()
```

```python
plt.figure(figsize=(12, 6))
players_team_drafted.plot(kind='bar', color='#7f7f7f')
plt.xlabel('Team Abbreviation')
plt.ylabel('Number of Players Drafted')
plt.title('Number of Players Drafted by Each Team')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Distribution of Players Drafted in 1st and 2nd Rounds

```python
first_round = df[df['draft_round'] == '1']
second_round = df[df['draft_round'] == '2']
```

```python
first_round_count = len(first_round)
second_round_count = len(second_round)
```

```python
labels = ['1st Round', '2nd Round']
sizes = [first_round_count, second_round_count]
colors = ['#7f7f7f', '#CD8363']
plt.figure(figsize=(18, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Players Drafted in 1st and 2nd Rounds', fontsize=14)
plt.tight_layout()
plt.show()
```

#### Height progress all throughput the years

```python
number_one_picks = df[df['draft_number'] == 1]
number_one_picks_list = number_one_picks[['season', 'player_name']]
number_one_picks_list = number_one_picks_list.sort_values('season')
print(number_one_picks_list)
```

```python
df['year'] = df['season'].str[:4].astype(int)
average_height_every_year = df.groupby('year')['player_height'].mean()
```

```python
plt.figure(figsize=(18, 6))
plt.plot(average_height_every_year.index, average_height_every_year.values, marker='o', linestyle='-', color='#7f7f7f')
plt.title('Average Height of NBA Players Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Height (cm)', fontsize=14)
plt.grid(True)
plt.show()
```

#### Most played games

```python
total_gp_by_player = df.groupby('player_name')['gp'].sum().reset_index()
total_gp_by_player_sorted = total_gp_by_player.sort_values(by='gp', ascending=False)
print(total_gp_by_player_sorted.head(10))
total_gp_by_player_sorted.to_csv('total_gp_by_player.csv', index=False)
```

```python
total_gp_by_player = df.groupby('player_name')['gp'].sum().reset_index()
total_gp_by_player_sorted = total_gp_by_player.sort_values(by='gp', ascending=False)
top_10_players = total_gp_by_player_sorted.head(10)
```

```python
plt.figure(figsize=(18, 6))
plt.bar(top_10_players['player_name'], top_10_players['gp'], color='#7f7f7f')
plt.title('Top 10 Players by Total Games Played', fontsize=16)
plt.xlabel('Player Name', fontsize=14)
plt.ylabel('Total Games Played (gp)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
